The following query gathers the names of each table, the count of columns (attributes), and the total number of rows in each table. It combines these results from various tables using the UNION ALL operator.

```python
%%sql
SELECT 
    'Customers' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Customers')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    customers

UNION ALL

SELECT 
    'Products' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Products')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    products

UNION ALL

SELECT 
    'ProductLines' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('ProductLines')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    productlines

UNION ALL

SELECT 
    'Orders' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Orders')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    orders

UNION ALL

SELECT 
    'OrderDetails' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('OrderDetails')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    orderdetails

UNION ALL

SELECT 
    'Payments' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Payments')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM
    payments

UNION ALL

SELECT 
    'Employees' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Employees')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    employees

UNION ALL

SELECT 
    'Offices' AS table_name,
    (SELECT COUNT(*) FROM pragma_table_info('Offices')) AS number_of_attributes,
    COUNT(*) AS number_of_rows
  FROM 
    offices;
```

### Task 1: Which Products Should We Order More of or Less of?
<a id="task-1-which-products-should-we-order-more-of-or-less-of"></a>

Low stock indicates the ratio of the total quantity ordered for each product to the quantity available in stock. We can focus on the ten highest ratios, identifying the top ten products that are nearly out of stock or completely depleted.

Product performance is defined as the total sales for each product.

Priority products for restocking are those that show high product performance and are close to being out of stock.
***
- Low stock = SUM(quantityOrdered) / quantityInStock

- Product performance = SUM(quantityOrdered * priceEach)

`Low stock`


```python
%%sql
WITH Low_Stock AS (
    SELECT p.productCode, 
           p.productName, 
           p.productLine,
           ROUND(SUM(od.quantityOrdered) * 1.0 / p.quantityInStock, 2) AS low_stock
      FROM products p
      JOIN orderdetails od ON p.productCode = od.productCode 
    GROUP BY p.productCode, p.productName, p.productLine

),
Product_Performance AS ( 
    SELECT p.productCode, 
           SUM(od.quantityOrdered * od.priceEach) AS product_performance
      FROM  orderdetails od
      JOIN products p ON od.productCode = p.productCode
    GROUP BY p.productCode
),
Priority_Products AS (
    SELECT productCode
      FROM Low_Stock
      WHERE low_stock < 0.5
      UNION ALL
    SELECT productCode
      FROM Product_Performance
)

SELECT  p.productCode, 
        p.productName, 
        p.productLine, 
        ls.low_stock
  FROM  Low_Stock ls
  JOIN products p  
    ON p.productCode = ls.productCode
 WHERE ls.productCode IN 
	              (SELECT productCode 
				     FROM Priority_Products)
ORDER BY  ls.low_stock DESC, p.productName
LIMIT 10;
```

```python
#convert query results into dataframe
query = """
     WITH Low_Stock AS (
    SELECT p.productCode, 
           p.productName, 
           p.productLine,
           ROUND(SUM(od.quantityOrdered) * 1.0 / p.quantityInStock, 2) AS low_stock
      FROM products p
      JOIN orderdetails od ON p.productCode = od.productCode 
    GROUP BY p.productCode, p.productName, p.productLine

),
Product_Performance AS ( 
    SELECT p.productCode, 
           SUM(od.quantityOrdered * od.priceEach) AS product_performance
      FROM  orderdetails od
      JOIN products p ON od.productCode = p.productCode
    GROUP BY p.productCode
),
Priority_Products AS (
    SELECT productCode
      FROM Low_Stock
      WHERE low_stock < 0.5
      UNION ALL
    SELECT productCode
      FROM Product_Performance
)

    SELECT  p.productCode, 
        p.productName, 
        p.productLine, 
        ls.low_stock
      FROM  Low_Stock ls
      JOIN products p  
        ON p.productCode = ls.productCode
      WHERE ls.productCode IN 
	              (SELECT productCode 
				     FROM Priority_Products)
     ORDER BY  ls.low_stock DESC, p.productName;
    
    """   
df1_total = pd.read_sql_query(query, conn)
   
```

### Task 2: How Should We Match Marketing and Communication Strategies to Customer Behavior?
<a id="task-2-how-should-we-match-marketing-and-communication-strategies-to-customer-behavior"></a>

This involves categorizing customers: finding the VIP (very important person) customers and those who are less engaged.

* VIP customers bring in the most profit for the store.

* Less-engaged customers bring in less profit.

Compute how much profit each customer generates:

```python
%%sql
    SELECT o.customerNumber, ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)),1) AS profit
    FROM orders o
    JOIN orderdetails od
        ON o.orderNumber = od.orderNumber
    JOIN products p
        ON p.productCode = od.productCode
    GROUP BY o.customerNumber
    ORDER BY profit DESC
    LIMIT 15;
```

 Finding the VIP Customers : The Top Five

```python
%%sql
    WITH profit_per_customer AS(
    SELECT o.customerNumber, ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)),2) AS profit
	  FROM orders o
	  JOIN orderdetails od
		ON o.orderNumber = od.orderNumber
	  JOIN products p
		ON p.productCode = od.productCode
	  GROUP BY o.customerNumber
	  ORDER BY profit DESC
)

    SELECT c.contactLastName, c.contactFirstName, c.city, c.country, ppc.profit
    FROM customers c
    JOIN profit_per_customer ppc
        ON c.customerNumber = ppc.customerNumber
    ORDER BY ppc.profit DESC
    LIMIT 5;
```

Finding Less Engaged Customers : The top five least-engaged customers


```python
%%sql
    WITH profit_per_customer AS(
    SELECT o.customerNumber, ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)),2) AS profit
	  FROM orders o
	  JOIN orderdetails od
		ON o.orderNumber = od.orderNumber
	  JOIN products p
		ON p.productCode = od.productCode
	  GROUP BY o.customerNumber
	  ORDER BY profit DESC
)

    SELECT c.contactLastName, c.contactFirstName, c.city, c.country, ppc.profit
    FROM customers c
    JOIN profit_per_customer ppc
        ON c.customerNumber = ppc.customerNumber
    ORDER BY ppc.profit ASC
    LIMIT 5;
```

### Task 3: How Much Can We Spend on Acquiring New Customers?
<a id="task-3-how-much-can-we-spend-on-acquiring-new-customers"></a>
To determine how much money we can spend acquiring new customers, we can compute the Customer Lifetime Value (LTV), which represents the average amount of money a customer generates. We can then determine how much we can spend on marketing.


Compute the average of customer profits:

```python
%%sql
    WITH profit_per_customer AS(
        SELECT o.customerNumber, SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)) AS profit
        FROM orders o
        JOIN orderdetails od
            ON o.orderNumber = od.orderNumber
        JOIN products p
            ON p.productCode = od.productCode
        GROUP BY o.customerNumber
        ORDER BY profit DESC
    )

    SELECT ROUND(AVG(profit),3) AS avg_profit
    FROM profit_per_customer;
```

LTV tells us how much profit an average customer generates during their lifetime with our store. We can use it to predict our future profit. So, if we get ten new customers next month, we'll earn 390,395 dollars, and we can decide based on this prediction how much we can spend on acquiring new customers.

### Task 4: Find the most profitable countries. <a class="anchor" id="Task_4"></a>
<a id="task-4-find-the-most-profitable-countries"></a>

```python
%%sql
    WITH profit_per_customer AS(
        SELECT o.customerNumber, ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)),2) AS profit
        FROM orders o
        JOIN orderdetails od
            ON o.orderNumber = od.orderNumber
        JOIN products p
            ON p.productCode = od.productCode
        GROUP BY o.customerNumber
        ORDER BY profit DESC
    )

    SELECT c.country, ROUND(SUM(ppc.profit),2) As total_profit
    FROM customers c
    JOIN profit_per_customer ppc
        ON c.customerNumber = ppc.customerNumber
    GROUP BY c.country
    ORDER BY total_profit DESC
    LIMIT 10;
```

```python
query_1 = """
    WITH profit_per_customer AS(
            SELECT o.customerNumber, ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)),2) AS profit
            FROM orders o
            JOIN orderdetails od
                ON o.orderNumber = od.orderNumber
            JOIN products p
                ON p.productCode = od.productCode
            GROUP BY o.customerNumber
            ORDER BY profit DESC
        )

        SELECT c.country, ROUND(SUM(ppc.profit),2) As total_profit
        FROM customers c
        JOIN profit_per_customer ppc
            ON c.customerNumber = ppc.customerNumber
        GROUP BY c.country
        ORDER BY total_profit DESC
        LIMIT 10;
 """
df_country = pd.read_sql_query(query_1, conn)
```

```python

fig = plt.figure(figsize=(10,5))
plt.bar(x=df_country['country'],height=df_country['total_profit'])
plt.grid(color='#95a5a6', linestyle='--', linewidth=0.5, axis='y', alpha=0.5)
plt.xticks(rotation=45)
plt.ylabel('Total Profit')
plt.title('The 10 Most Profitable Countries')

#remove all borders
for spine in plt.gca().spines.values():
    spine.set_visible(False)

#set left and bottom borders to visible
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

#display y-axis values with commas
formatter = mticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x))
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

 ### Task 7: Employee Productivity: Analyze the total sales generated by each employee.
<a id="task-7-employee-productivity-analyze-the-total-sales-generated-by-each-employee"></a>
We will track the total sales attributed to each employee by calculating the number of orders they processed and the total revenue generated from these orders.

```python
%%sql
    SELECT e.firstName, e.lastName, 
            COUNT(o.orderNumber) AS total_orders,
            ROUND(SUM(od.quantityOrdered * od.priceEach),2) AS total_sales
        FROM employees e
        JOIN customers c
        ON e.employeeNumber = c.salesRepEmployeeNumber
        JOIN orders o
        ON o.customerNumber = c.customerNumber
        JOIN orderdetails od
        ON o.orderNumber = od.orderNumber
        GROUP BY e.employeeNumber
        ORDER BY total_sales DESC;
```

* Gerard Hernandez and Leslie Jennings are the top two employees in terms of total orders and sales, with over $1.2 million and $1.08 million in sales, respectively.
* The first five employees have all generated sales exceeding $700,000, suggesting a high level of productivity and effectiveness in their sales efforts.



### Task 8: Identify the office with the highest sale.
<a id="task-8-identify-the-office-with-the-highest-sales"></a>

```python
%%sql
    WITH EmployeeSales AS (
    SELECT e.firstName, e.lastName, e.officeCode,
            COUNT(o.orderNumber) AS total_orders,
            ROUND(SUM(od.quantityOrdered * od.priceEach),2) AS total_sales
        FROM employees e
        JOIN customers c
        ON e.employeeNumber = c.salesRepEmployeeNumber
        JOIN orders o
        ON o.customerNumber = c.customerNumber
        JOIN orderdetails od
        ON o.orderNumber = od.orderNumber
        GROUP BY e.employeeNumber,e.officeCode
        )
    SELECT o.officeCode, o.city, o.country,
        ROUND(SUM(es.total_sales),2) AS office_total_sales
    FROM offices o
    JOIN EmployeeSales es
        ON o.officeCode = es.officeCode
    GROUP BY o.officeCode, o.city, o.state
    ORDER BY office_total_sales DESC;
```
